# weight 

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add an edge with a weight
G.add_edge(1, 2, weight=12.5)

# Set labels with font weight
labels = {node: str(node) for node in G.nodes()}
labels[1] = r'$\mathbf{1}$'  # Make label '1' bold using LaTeX math notation

# Set edge labels
edge_labels = {(u, v): str(attr['weight']) for u, v, attr in G.edges(data=True)}

# Visualization
pos = nx.spring_layout(G)

# Draw the graph with custom labels and font properties
nx.draw(G, pos, with_labels=True, labels=labels, font_weight='bold')

# Draw edge labels
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Show the graph
plt.show()
```

# Accessing edges and neighbors

```python
G.add_edge(1, 3)
G[1][3]['color'] = "blue"
```

```python
G = nx.Graph()
G.add_weighted_edges_from([(1, 2, 0.125), (1, 3, 0.75), (2, 4, 1.2), (3, 4, 0.375)])
for n, nbrs in G.adj.items():
   for nbr, eattr in nbrs.items():
       wt = eattr['weight']
       if wt < 0.5: print(f"({n}, {nbr}, {wt:.3})")
```

```python
for (u, v, wt) in G.edges.data('weight'):
    if wt < 0.5:
        print(f"({u}, {v}, {wt:.3})")
```
